Working with ChEMBL data using the PostgreSQL Cartridge
====

Building the database
----

Start by downloading the PostgreSQL dump of the ChEMBL database from ftp://ftp.ebi.ac.uk/pub/databases/chembl/ChEMBLdb/latest


Now run `psql chembl_14` (or whatever the correct version is) and execute the following commands. *Note:* This will take a while and will generate a fair amount of output.

    create extension rdkit;
    create schema rdk;
    select molregno,mol_from_ctab(molfile::cstring) m into temporary table rdk.mols from compound_structures;
    select * into rdk.mols from tmols where m is not null;
    create index molidx on rdk.mols using gist(m);
    select molregno,morganbv_fp(m,2) as mfp2,morganbv_fp(m,3) as mfp3,featmorganbv_fp(m,2) as ffp2 into rdk.fps from rdk.mols;
    CREATE INDEX mfp2idx on rdk.fps using gist(mfp2);
    create index mfp3idx on rdk.fps using gist(mfp3);
    create index ffp2idx on rdk.fps using gist(ffp2);
    SELECT molregno, atompairbv_fp(m) as pairfp,torsionbv_fp(m) AS torsionfp INTO rdk.pairbvfps FROM rdk.mols;
    CREATE INDEX apfpbvidx ON rdk.pairbvfps USING gist (pairfp);
    CREATE INDEX torsionfpbvidx ON rdk.pairbvfps USING gist (torsionfp);
    select molregno,torsion_fp(m) as torsionfp,atompair_fp(m) as pairfp into rdk.pairfps from rdk.mols;
    create index apfpidx on rdk.pairfps using gist(pairfp sfp_low_ops);
    create index torsionfpidx on rdk.pairfps using gist (torsionfp sfp_low_ops);

    CREATE OR REPLACE FUNCTION mfp2_query(m mol)
    RETURNS TABLE (molregno numeric, similarity double precision) AS
    $BODY$
        SELECT molregno,
             -- Tanimoto similarity
             tanimoto_sml(mfp2,morganbv_fp($1,2)) similarity
        FROM rdk.fps
             -- Boolean operator returning true if the Tanimoto similarity
             -- is above the user-defined limit
        WHERE mfp2 % morganbv_fp($1,2)
             -- KNN-GIST order by operator for Tanimoto similarity
        ORDER BY mfp2 <%> morganbv_fp($1,2) 
    $BODY$
    LANGUAGE sql IMMUTABLE STRICT;

    CREATE OR REPLACE FUNCTION mfp2_query(smiles TEXT)
    RETURNS TABLE (molregno numeric, similarity double precision) AS
    $BODY$
         SELECT mfp2_query($1::mol);
    $BODY$
    LANGUAGE sql IMMUTABLE STRICT;

    CREATE OR REPLACE FUNCTION mfp3_query(smiles TEXT)
    RETURNS TABLE (molregno numeric, similarity double precision) AS
    $BODY$
      SELECT molregno,
             -- Tanimoto similarity
             tanimoto_sml(mfp3,morganbv_fp($1::mol,3)) similarity
        FROM rdk.fps
             -- Boolean operator returning true if the Tanimoto similarity
             -- is above the user-defined limit
        WHERE mfp3 % morganbv_fp($1::mol,3)
             -- KNN-GIST order by operator for Tanimoto similarity
        ORDER BY mfp3 <%> morganbv_fp($1::mol,3) 
    $BODY$
    LANGUAGE sql IMMUTABLE STRICT;

    CREATE OR REPLACE FUNCTION ffp2_query(smiles TEXT)
    RETURNS TABLE (molregno numeric, similarity double precision) AS
    $BODY$
      SELECT molregno,
             -- Tanimoto similarity
             tanimoto_sml(ffp2,featmorganbv_fp($1::mol,2)) similarity
        FROM rdk.fps
             -- Boolean operator returning true if the Tanimoto similarity
             -- is above the user-defined limit
        WHERE ffp2 % featmorganbv_fp($1::mol,2)
             -- KNN-GIST order by operator for Tanimoto similarity
        ORDER BY ffp2 <%> featmorganbv_fp($1::mol,2) 
    $BODY$
    LANGUAGE sql IMMUTABLE STRICT;

    CREATE OR REPLACE FUNCTION torsionbv_query(smiles TEXT)
    RETURNS TABLE (molregno numeric, similarity double precision) AS
    $BODY$
      SELECT molregno,
             -- Tanimoto similarity
             tanimoto_sml(torsionfp,torsionbv_fp($1::mol)) similarity
      FROM rdk.pairbvfps
             -- Boolean operator returning true if the Tanimoto similarity
             -- is above the user-defined limit
      WHERE torsionfp % torsionbv_fp($1::mol)
         -- KNN-GIST order by operator for Tanimoto similarity
      ORDER BY torsionfp <%> torsionbv_fp($1::mol) 
    $BODY$
    LANGUAGE sql IMMUTABLE STRICT;


```python
from rdkit import Chem
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem import Draw
```

```python
import psycopg2
conn=psycopg2.connect(dbname="chembl_14",user="db_user_r",password="smiles")
curs = conn.cursor()
```

Start with some substructure and similarity queries

```python
Chem.MolFromSmiles('O=C1OC2=C(C=C1)C1=C(C=CCO1)C=C2')
```

```python
conn.rollback()
curs.execute("select molregno,m smiles from rdk.mols where m @> 'O=C1OC2=C(C=C1)C1=C(C=CCO1)C=C2'")
d = curs.fetchall()
sss_results = dict(d)
Draw.MolsToGridImage([Chem.MolFromSmiles(y) for x,y in d[:40]],molsPerRow=4,legends=[str(x) for x,y in d[:40]])
```

```python
conn.rollback()
curs.execute("set rdkit.tanimoto_threshold=0.6")
curs.execute("select molregno,m smiles,similarity from mfp2_query(%s) join rdk.mols using (molregno) order by similarity desc",(sss_results[440381],))
d = curs.fetchall()
Draw.MolsToGridImage([Chem.MolFromSmiles(y) for x,y,z in d],molsPerRow=4,legends=['%s: %.2f'%(x,z) for x,y,z in d])
```

```python
conn.rollback()
curs.execute("select molregno,m smiles,similarity from ffp2_query(%s) join rdk.mols using (molregno) order by similarity desc",(sss_results[440381],))
d = curs.fetchall()
Draw.MolsToGridImage([Chem.MolFromSmiles(y) for x,y,z in d],molsPerRow=4,legends=['%s: %.2f'%(x,z) for x,y,z in d])
```

##Taking advantage of psycopg2 extensions

```python
conn.rollback()
curs.execute("SELECT 'c1ccccc1'::mol")
mol_oid=curs.description[0].type_code
def cast_mol(val,curs):
    if val is None: return None
    m = Chem.MolFromSmiles(str(val))
    return m
pgMol=psycopg2.extensions.new_type((mol_oid,),"pgMol",cast_mol)
psycopg2.extensions.register_type(pgMol)
def adapt_mol(mol):
    return psycopg2.extensions.AsIs("'%s'"%Chem.MolToSmiles(mol,True))
psycopg2.extensions.register_adapter(Chem.Mol,adapt_mol)
```
