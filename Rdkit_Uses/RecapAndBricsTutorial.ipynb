```python
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem import Draw
from rdkit.Chem import Recap,BRICS
```

```python
cdk2mols = [x for x in Chem.SDMolSupplier('data/cdk2.sdf') if x is not None]
for m in cdk2mols: AllChem.Compute2DCoords(m)
```

BRICS
====

```python
m1 = cdk2mols[0]
m1
```

```python
pieces = BRICS.BRICSDecompose(m1)
```

```python
pieces
```

```python
cdk2mols[20]
```

```python
pieces = BRICS.BRICSDecompose(cdk2mols[20])
pieces
```

```python
allfrags=set()
for m in cdk2mols:
    pieces = BRICS.BRICSDecompose(m)
    allfrags.update(pieces)
allfrags
```

```python
len(allfrags)
```

```python
len(cdk2mols)
```

If you are interested in the final pieces from a BRICS decomposition, it is more efficient to use `BRICS.BreakBRICSBonds()`. This doesn't do as much book-keeping work as `BRICS.BRICSDecompose()`, so it's considerably faster.

```python
m = Chem.Mol(cdk2mols[0].ToBinary())
m
```

```python
fragmented=BRICS.BreakBRICSBonds(m)
pieces=Chem.GetMolFrags(fragmented,asMols=True)
Draw.MolsToGridImage(pieces)
```

RECAP
====

Here the implementation behaves somewhat differently: it returns a hierarchy of fragments.

```python
m=cdk2mols[35]
m
```

```python
pieces = Recap.RecapDecompose(m)
pieces
```

```python
pieces.children
```

```python
Chem.MolFromSmiles('[*]c1nc2c(ncn2C(C)C)c(Nc2ccc(C(=O)[O-])c(Cl)c2)n1')
```

```python
pieces.children['[*]c1nc2c(ncn2C(C)C)c(Nc2ccc(C(=O)[O-])c(Cl)c2)n1'].children
```

Nodes also have parents.
Notice that the hierarchy is a network, not a tree -- each node can have multiple children and multiple parents

```python
pieces.children['[*]C(C)C'].parents
```

We can short-cut the hierarchy traversal with the `GetLeaves()` method:

```python
pieces.GetLeaves()
```
